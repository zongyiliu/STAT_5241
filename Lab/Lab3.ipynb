```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy.linalg import inv, eig
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from statsmodels.api import OLS
```

```python
ozone = pd.read_csv('ozone.csv')
```

**Exploring the data through scatterplots and histograms**

```python
# Number of variables
variables = ozone.columns
num_vars = len(variables)

# Create a figure and axis grid
fig, axes = plt.subplots(num_vars, num_vars, figsize=(10, 10))

# Generate plots
for i, var1 in enumerate(variables):
    for j, var2 in enumerate(variables):
        ax = axes[i, j]
        if i != j:
            ax.scatter(ozone[var2], ozone[var1], s=5)
        else:
            # For diagonal, display histogram
            ax.hist(ozone[var1], bins=15, color='gray')
        if i == num_vars - 1:
            ax.set_xlabel(var2)
        if j == 0:
            ax.set_ylabel(var1)
        ax.tick_params(axis='both', which='both', labelsize=6)

# Adjust layout
plt.tight_layout()
plt.show()
```

**Least squares using LinearRegression()**

```python
# Least squares using LinearRegression()
X = ozone.drop(columns=['ozone']).to_numpy()
Y = ozone['ozone'].to_numpy()
model = LinearRegression()
model.fit(X,Y)

print("The coefficients are "+str(model.coef_)+".")
print("The intercept is "+str(model.intercept_)+".")
```

```python
# Least squares using statsmodels
lsmodel = OLS(Y,X).fit()
lsmodel.summary()
```

**Directly calculating least squares**

```python
# Directly calculating least squares

# Adding ones so that betahat includes intercept
aX = np.hstack([np.ones((X.shape[0], 1)), X])

# Estimated coefficients
betahat = inv(aX.T @ aX) @ (aX.T @ Y)

# Hat matrix
H = aX @ inv(aX.T @ aX) @ aX.T
eigenvalues = eig(H)[0]
# Round up to 12 decimals to account for numerical precision
np.round(eigenvalues[0:15], 12)
```

```python
# Fitted data
Yhat = aX @ betahat
Yhat = H @ Y

# Compare to LinearRegression()
np.column_stack([betahat, np.concatenate(([model.intercept_], model.coef_))])
```

**Checking intercepts: least squares with centered and standardized data**

```python
# Center y
Y_centered = Y - np.mean(Y)

# Center X (not scaled)
X_centered = X - np.mean(X, axis=0)

# Standardize X (centered and scaled)
scaler = StandardScaler(with_mean=True, with_std=True)
X_standardized = scaler.fit_transform(X)
```

```python
# Fit the original linear model
model = LinearRegression()
model.fit(X, Y)
yhat = model.predict(X)

# Fit the centered model
model_centered = LinearRegression(fit_intercept=False)
model_centered.fit(X_centered, Y_centered)
yhat_centered = model_centered.predict(X_centered) + np.mean(Y)
print("Sum of differences (centered vs. original):", np.sum(yhat_centered - yhat))

# Fit the standardized model
model_standardized = LinearRegression(fit_intercept=False)
model_standardized.fit(X_standardized, Y_centered)
yhat_standardized = model_standardized.predict(X_standardized) + np.mean(Y)
print("Sum of differences (standardized vs. original):", np.sum(yhat_standardized - yhat))
```
